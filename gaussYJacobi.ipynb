```python
import matplotlib.pyplot as plt

def eliminacion_gaussiana(A_b):
    # Implementar aquí la eliminación gaussiana si es necesario
    pass

def matriz_aumentada(A, b):
    return [row + [b_i] for row, b_i in zip(A, b)]

def gauss_jacobi(A, b, x0, tol=1e-5, max_iter=1000):
    n = len(A)
    x = x0.copy()
    traj = [x.copy()]
    for _ in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            sum_ax = sum(A[i][j] * x[j] for j in range(n) if i != j)
            x_new[i] = (b[i] - sum_ax) / A[i][i]
        traj.append(x_new.copy())
        if max(abs(x_new[i] - x[i]) for i in range(n)) < tol:
            break
        x = x_new
    return x, traj

def gauss_seidel(A, b, x0, tol=1e-5, max_iter=1000):
    n = len(A)
    x = x0.copy()
    traj = [x.copy()]
    for _ in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            sum_ax = sum(A[i][j] * x_new[j] for j in range(n) if i != j)
            x_new[i] = (b[i] - sum_ax) / A[i][i]
        traj.append(x_new.copy())
        if max(abs(x_new[i] - x[i]) for i in range(n)) < tol:
            break
        x = x_new
    return x, traj

# Dibujar la trayectoria
def plot_trajectory(traj_jacobi, traj_seidel):
    plt.figure(figsize=(10, 5))

    # Trajectory of Gauss-Jacobi
    plt.subplot(1, 2, 1)
    plt.plot(*zip(*traj_jacobi), marker='o', label='Gauss-Jacobi')
    plt.title('Gauss-Jacobi Trajectory')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid(True)

    # Trajectory of Gauss-Seidel
    plt.subplot(1, 2, 2)
    plt.plot(*zip(*traj_seidel), marker='o', label='Gauss-Seidel')
    plt.title('Gauss-Seidel Trajectory')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Parámetros iniciales
A = [[1, 1], [-2, 5]]
b = [7, 0]
x0 = [1, 1]

# Ejecutar métodos
sol_jacobi, traj_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=1e-5, max_iter=1000)
sol_seidel, traj_seidel = gauss_seidel(A=A, b=b, x0=x0, tol=1e-5, max_iter=1000)

# Graficar las trayectorias
plot_trajectory(traj_jacobi, traj_seidel)

# Imprimir soluciones finales
print(f'Solución Gauss-Jacobi: {sol_jacobi}')
print(f'Solución Gauss-Seidel: {sol_seidel}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def gauss_jacobi(A, b, x0, tol=1e-5, max_iter=1000):
    n = len(A)
    x = np.array(x0, dtype=float)
    traj = [x.copy()]
    for _ in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            sum_ax = sum(A[i][j] * x[j] for j in range(n) if i != j)
            x_new[i] = (b[i] - sum_ax) / A[i][i]
        traj.append(x_new.copy())
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            break
        x = x_new
    return x, traj

def gauss_seidel(A, b, x0, tol=1e-5, max_iter=1000):
    n = len(A)
    x = np.array(x0, dtype=float)
    traj = [x.copy()]
    for _ in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            sum_ax = sum(A[i][j] * x_new[j] for j in range(n) if i != j)
            x_new[i] = (b[i] - sum_ax) / A[i][i]
        traj.append(x_new.copy())
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            break
        x = x_new
    return x, traj

def plot_trajectory(traj_jacobi, traj_seidel):
    plt.figure(figsize=(10, 5))

    # Trajectory of Gauss-Jacobi
    plt.subplot(1, 2, 1)
    plt.plot(*zip(*traj_jacobi), marker='o', label='Gauss-Jacobi')
    plt.title('Gauss-Jacobi Trajectory')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid(True)

    # Trajectory of Gauss-Seidel
    plt.subplot(1, 2, 2)
    plt.plot(*zip(*traj_seidel), marker='o', label='Gauss-Seidel')
    plt.title('Gauss-Seidel Trajectory')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Parámetros iniciales para el nuevo sistema de ecuaciones
A = [[1, 1], [-2, 1]]
b = [6, 0]
x0 = [1, 1]  # Estimación inicial

# Ejecutar métodos
sol_jacobi, traj_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=1e-5, max_iter=1000)
sol_seidel, traj_seidel = gauss_seidel(A=A, b=b, x0=x0, tol=1e-5, max_iter=1000)

# Graficar las trayectorias
plot_trajectory(traj_jacobi, traj_seidel)

# Imprimir soluciones finales
print(f'Solución Gauss-Jacobi: {sol_jacobi}')
print(f'Solución Gauss-Seidel: {sol_seidel}')
```
